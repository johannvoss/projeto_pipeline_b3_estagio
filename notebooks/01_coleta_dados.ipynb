## 1. Importação das bibliotecas

```python
import pandas as pd
import yfinance as yf
```

---

## 2. Definição dos parâmetros de coleta

Antes de sair baixando dados, vale centralizar os parâmetros em variáveis. Assim fica fácil ajustar o período ou adicionar novos ativos no futuro sem precisar caçar o valor espalhado pelo código.

```python
# Tickers no formato que o Yahoo Finance entende para ações brasileiras
lista_tickers = ["PETR4.SA", "VALE3.SA", "ITUB4.SA"]

# Período dos dados — últimos 12 meses
data_inicio = "2024-01-01"
data_fim = "2025-01-01"

print(f"Ativos selecionados: {lista_tickers}")
print(f"Período: {data_inicio} até {data_fim}")
```

---

## 3. Coleta dos dados

Vamos baixar os dados de cada ativo separadamente. Preferi fazer assim (em vez de usar o `download` em lote do yfinance) para ter mais controle sobre cada DataFrame e facilitar a leitura do código.

```python
# Baixando os dados da Petrobras
dados_petr4 = yf.download("PETR4.SA", start=data_inicio, end=data_fim)

# Baixando os dados da Vale
dados_vale3 = yf.download("VALE3.SA", start=data_inicio, end=data_fim)

# Baixando os dados do Itaú
dados_itub4 = yf.download("ITUB4.SA", start=data_inicio, end=data_fim)

print("✅ Dados coletados com sucesso!")
```

## 5. Salvamento dos dados brutos

Agora vamos salvar os três DataFrames como arquivos CSV na pasta `data/raw/`. A ideia de manter os dados brutos separados dos dados tratados é uma boa prática: se algo der errado no tratamento, a gente sempre pode voltar à fonte sem precisar baixar tudo de novo.

```python
import os

# Garantindo que a pasta existe antes de tentar salvar
os.makedirs("../data/raw", exist_ok=True)

# Caminhos de destino para cada arquivo
caminho_arquivo_petr4_bruto = "../data/raw/petr4_dados_brutos.csv"
caminho_arquivo_vale3_bruto = "../data/raw/vale3_dados_brutos.csv"
caminho_arquivo_itub4_bruto = "../data/raw/itub4_dados_brutos.csv"

# Salvando cada ativo em seu respectivo arquivo
dados_petr4.to_csv(caminho_arquivo_petr4_bruto)
dados_vale3.to_csv(caminho_arquivo_vale3_bruto)
dados_itub4.to_csv(caminho_arquivo_itub4_bruto)

print("✅ Arquivos salvos com sucesso em data/raw/:")
print(f"   → {caminho_arquivo_petr4_bruto}")
print(f"   → {caminho_arquivo_vale3_bruto}")
print(f"   → {caminho_arquivo_itub4_bruto}")
```
